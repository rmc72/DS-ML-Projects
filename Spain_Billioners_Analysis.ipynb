# Spain Billionaires Analysis with Python

The dataset that I am using to analyze the data about billionaires around the world was curated by Forbes and is downloaded from Kaggle. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#dataset can be downloaded from https://www.kaggle.com/datasets/roysouravcu/forbes-billionaires-of-2021
data = pd.read_csv("Billionaire.csv")
data.head()
```

Let's select billionaires from Spain

```python
data = data[data['Country']=='Spain']
data
```

```python
data.shape[0]
```

There are 30 Spain billionaires in Forbes list in 2021

Now letâ€™s have a look at the top 10 Spain billionaires according to their NetWorth:

```python
df = data.sort_values(by = ["NetWorth"], ascending=False).head(10)
plt.figure(figsize=(20, 10))
sns.histplot(x="Name", hue="NetWorth", data=df)
plt.show()
```

Now letâ€™s have a look at the top 5 Spain domains with the most number of billionaires:

```python
a = data["Source"].value_counts().head()
index = a.index
sources = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(5, 5))
plt.pie(sources, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Domains to Become a Billionaire in Spain", fontsize=20)
plt.show()
```

Now letâ€™s have a look at the top 5 Spain industries with the most number of billionaires:

```python

a = data["Industry"].value_counts().head()
index = a.index
industries = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(5, 5))
plt.pie(industries, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Industries with Most Number of Billionaires in Spain", fontsize=20)
plt.show()
```
